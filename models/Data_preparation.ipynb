남구 65세이상 기초수급자 거주지 시각화

```python
pd.set_option('display.float_format', '{:.10f}'.format)
shp_file_path_3 = "Data\남구 65세 이상 기초수급자 거주지(통신사비율보정) - 복사본\남구 65세이상 기초수급자(통산사비율보정).shp"
df3 = gpd.read_file(shp_file_path_3, encoding='cp949')
df3.head()
```

```python
# 가중치 설정
weight_demand = 0.5
weight_price = 0.5

# 집값 기반 점수 계산 (집값이 낮을수록 높은 점수)
df3['price_score'] = 1 / df3['aver_price']

# 최종 우선순위 점수 계산
df3['priority_score'] = weight_demand * df3['demand'] + weight_price * df3['price_score']

# 결과 출력
df3.head()
```

```python
# pip install openrouteservice
```

```python
import folium
import osmnx as ox
import networkx as nx
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import branca.colormap as cm

# df3의 데이터를 지도에 CircleMarker로 표시
df3['coords'] = df3['geometry'].apply(lambda x: (x.y, x.x))

# 가격 데이터를 바탕으로 색상/투명도 그라데이션 적용
scaler = MinMaxScaler()
df3['scaled_price'] = scaler.fit_transform(df3[['aver_price']])

# 컬러맵 생성 (집 값이 저렴할수록 빨간색, 비쌀수록 초록색)
colormap = cm.LinearColormap(colors=['red', 'yellow', 'green'], vmin=0, vmax=1)

# folium 맵 생성 (타일셋을 영어 지명으로 설정)
center = [35.530, 129.320]  # 중심 좌표
m = folium.Map(location=center, zoom_start=14, tiles="OpenStreetMap")

# df3의 데이터를 CircleMarker로 지도에 추가 (가격에 따른 그라데이션)
for idx, row in df3.iterrows():
    color = colormap(row['scaled_price'])
    folium.CircleMarker(
        location=row['coords'],
        radius=4,  # 반경 설정
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=f"Price: {row['aver_price']}"
    ).add_to(m)

# 도로 네트워크 데이터를 가져오기
G = ox.graph_from_point(center, dist=4000, network_type='drive', simplify=True)

# 첫 번째 셔틀 경로 좌표 설정
coordinates_with_names_1 = [
    (35.5299, 129.3299, "복지관 주차장"),
    (35.5266, 129.3268, "대현동 행정복지센터길 건너편 버스정류장"),
    (35.5288, 129.3194, "수암시장 (울산중앙약국)"),
    (35.5309, 129.3145, "동서오거리 정류장"),
    (35.532, 129.3106, "김한의원"),
    (35.5331, 129.3034, "대흥교회 버스정류장 (젠피아양복점)"),
    (35.5339, 129.2966, "새현대의원"),
    (35.5374, 129.3059, "산업은행"),
    (35.5417, 129.3063, "리틀쥬동물의료원 (뚜레쥬르)"),
    (35.5477, 129.308, "백정형외과ㆍ신경외과"),
    (35.5472, 129.309, "바른 눈안경 (태화로타리) 앞"),
    (35.5438, 129.3105, "신정시장 (신정과일나라)"),
    (35.5397, 129.3121, "시청 (리무진 버스정류장)"),
    (35.5346, 129.317, "롯데마트"),
    (35.5353, 129.3199, "굿모닝병원"),
    (35.5364, 129.3238, "달동현대 버스정류장"),
    (35.5356, 129.3284, "경남은행"),
    (35.5299, 129.3299, "복지관 버스정류장")
]

# 두 번째 셔틀 경로 좌표 설정
coordinates_with_names_2 = [
    (35.5299, 129.3299, "복지관 버스정류장"),
    (35.5243, 129.3312, "야음시장 (DC마트 앞)"),
    (35.5216, 129.3355, "번개시장 (휴대폰 도매할인마트)"),
    (35.5179, 129.3407, "변전소"),
    (35.5204, 129.3385, "야음장생포 행정복지센터 (수암로 버스정류장)"),
    (35.5237, 129.3373, "남구구민체육센터 농협은행 버스정류장"),
    (35.5245, 129.3317, "야음시장"),
    (35.5299, 129.3299, "복지관 주차장")
]

# 첫 번째 경로 생성 및 폴리라인 추가
for i in range(len(coordinates_with_names_1) - 1):
    start_point = coordinates_with_names_1[i][:2]
    end_point = coordinates_with_names_1[i + 1][:2]
    
    # 최단 경로 계산
    start_node = ox.distance.nearest_nodes(G, start_point[1], start_point[0])
    end_node = ox.distance.nearest_nodes(G, end_point[1], end_point[0])
    route = nx.shortest_path(G, start_node, end_node, weight='length')
    route_coords = [(G.nodes[node]['y'], G.nodes[node]['x']) for node in route]
    
    # 경로를 지도에 추가 (파란색)
    folium.PolyLine(locations=route_coords, color="blue").add_to(m)

# 두 번째 경로 생성 및 폴리라인 추가
for i in range(len(coordinates_with_names_2) - 1):
    start_point = coordinates_with_names_2[i][:2]
    end_point = coordinates_with_names_2[i + 1][:2]
    
    # 최단 경로 계산
    start_node = ox.distance.nearest_nodes(G, start_point[1], start_point[0])
    end_node = ox.distance.nearest_nodes(G, end_point[1], end_point[0])
    route = nx.shortest_path(G, start_node, end_node, weight='length')
    route_coords = [(G.nodes[node]['y'], G.nodes[node]['x']) for node in route]
    
    # 경로를 지도에 추가 (빨간색)
    folium.PolyLine(locations=route_coords, color="red").add_to(m)

# 경로에 마커 추가 및 팝업 추가 (첫 번째 경로)
for lat, lon, name in coordinates_with_names_1:
    folium.Marker(
        location=[lat, lon],
        icon=folium.Icon(color='blue',icon_size=(40, 40)),
        popup=name  # 각 지점의 이름을 팝업으로 표시
    ).add_to(m)

# 경로에 마커 추가 및 팝업 추가 (두 번째 경로)
for lat, lon, name in coordinates_with_names_2:
    folium.Marker(
        location=[lat, lon],
        icon=folium.Icon(color='red',icon_size=(40, 40)),
        popup=name  # 각 지점의 이름을 팝업으로 표시
    ).add_to(m)

# 색상 범례 추가
colormap.add_to(m)

# 지도 출력
m
```
